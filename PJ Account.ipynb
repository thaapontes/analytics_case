```python
import pandas as pd
```

```python
#Loading datasets
customers = pd.read_csv("customers.csv", sep = ',', low_memory = False)
companies = pd.read_csv("companies.csv", sep = ',', low_memory = False)
customers.shape # 44, 7
companies.shape #50, 7
```

```python
#Join between cutomers and companies
df_inner = pd.merge(customers, companies, on='cpf', how='inner')
df_inner.shape #46, 13
df_inner.head()
```

```python
#Companies that are going to receive the invite
invite = df_inner[['cpf','cnpj', 'name', 'email_address']].groupby(['cnpj', 'name', 'email_address']).agg(['count'])
invite.columns = invite.columns.get_level_values(0)
invite = invite.sort_values(by=['cpf']).reset_index()
invite = invite.where(invite['cpf'] == 1).dropna()
invite.shape # 34, 4
invite.head()
```

```python
#create invites csv
invite.to_csv('companies_invites.csv')
```

```python
#dataset for tableau
df_right = pd.merge(customers, companies, on='cpf', how='right')
count_mei = df_inner[['cpf','cnpj']].groupby(['cnpj']).agg(['count'])
df_left = pd.merge(df_right, count_mei, on='cnpj', how='left')
df_left.to_csv('all_companies.csv')
df_left.head()
```
